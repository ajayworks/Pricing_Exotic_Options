# Monte Carlo Greeks – Pathwise and Likelihood Ratio Methods

This notebook implements two techniques for computing option Greeks via Monte Carlo simulation:
- **Pathwise Derivative Method** (exact derivative of payoff wrt parameters)
- **Likelihood Ratio Method** (shifts derivative to probability density)

We compute:
- **Delta** via both methods
- **Vega** via Likelihood Ratio
- And compare against finite difference benchmarks

```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys

sys.path.append(os.path.abspath("../../"))

from pricing.bsm import black_scholes_price 
```

```python
def simulate_gbm_paths(S0, T, r, sigma, n_paths):
    Z = np.random.normal(0, 1, n_paths)
    ST = S0 * np.exp((r - 0.5 * sigma ** 2) * T + sigma * np.sqrt(T) * Z)
    return ST, Z
```

```python
def mc_pathwise_delta(S0, K, T, r, sigma, n_paths=100000):
    ST, Z = simulate_gbm_paths(S0, T, r, sigma, n_paths)
    payoff_deriv = np.where(ST > K, ST / S0, 0.0)  # ∂Payoff/∂S0
    discounted = np.exp(-r * T) * payoff_deriv
    return np.mean(discounted)
```

```python
def mc_lr_greeks(S0, K, T, r, sigma, n_paths=100000):
    ST, Z = simulate_gbm_paths(S0, T, r, sigma, n_paths)
    payoff = np.maximum(ST - K, 0)

    lr_delta = payoff * Z / (S0 * sigma * np.sqrt(T))
    lr_vega = payoff * (Z**2 - 1) / sigma

    discount = np.exp(-r * T)
    return discount * np.mean(lr_delta), discount * np.mean(lr_vega)
```

```python
# Parameters
S0, K, T = 100, 100, 1.0
r, sigma = 0.05, 0.2
n_paths = 100000

# Monte Carlo Greeks
delta_pw = mc_pathwise_delta(S0, K, T, r, sigma, n_paths)
delta_lr, vega_lr = mc_lr_greeks(S0, K, T, r, sigma, n_paths)

# Finite diff for reference
def bs_delta(S, K, T, r, sigma):
    from scipy.stats import norm
    d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
    return norm.cdf(d1)

print(f"MC Pathwise Delta:     {delta_pw:.4f}")
print(f"MC Likelihood Delta:   {delta_lr:.4f}")
print(f"MC Likelihood Vega:    {vega_lr:.4f}")
print(f"BS Analytical Delta:   {bs_delta(S0, K, T, r, sigma):.4f}")
```

```python
path_sizes = np.arange(1000, 50001, 2000)
delta_vals = [mc_pathwise_delta(S0, K, T, r, sigma, n) for n in path_sizes]

plt.figure(figsize=(8, 5))
plt.plot(path_sizes, delta_vals, marker='o', label='MC Pathwise Delta')
plt.axhline(bs_delta(S0, K, T, r, sigma), color='red', linestyle='--', label='Analytical')
plt.xlabel("Number of Paths")
plt.ylabel("Delta")
plt.title("Convergence of Delta (Pathwise)")
plt.legend()
plt.grid(True)
plt.show()
```

## Summary

- **Pathwise method** provides accurate Delta for differentiable payoffs like calls.
- **Likelihood Ratio** extends to both Delta and Vega.
- Monte Carlo Greeks converge slower but are model-flexible, supporting exotic payoffs.
