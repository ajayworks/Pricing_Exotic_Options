```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulate stock prices of multiple paths at maturity

def simulate_gbm_paths(S0, T, r, sigma, N):
    Z = np.random.normal(0, 1, N)

    S_T = S0 * np.exp((r - 0.5 * sigma**2) * T + (sigma * np.sqrt(T) * Z))

    return S_T
```

```python
S_T = simulate_gbm_paths(100, 1, 0.05, 0.2, 1000)
```

```python
# Calculate payoff at S_T

def calculate_payoff(S_T, K, option_type):
        
    if option_type.lower() == 'call':
        payoff = np.maximum((S_T - K), 0)
        
    elif option_type.lower() == 'put':
        payoff = np.maximum((K - S_T), 0)

    else:
        raise ValueError("Invalid input: Enter 'call' or 'put'")

    return payoff
```

```python
payoff = calculate_payoff(S_T, 100, 'put')
```

```python
# Monte Carlo estimate of the option price

def discount_payoff(payoff, r, T):
    option_price = np.exp(-r * T) * np.mean(payoff)
    return option_price

# Example usage
price = discount_payoff(payoff, 0.05, 1)
print(f"Estimated Option Price: {price:.4f}")
```

```python
def plot_histograms(S_T, payoff, option_type):
    plt.figure(figsize = (10,5))
    plt.hist(S_T, bins = 50, color = 'skyblue', edgecolor = 'black')
    plt.title("Simulated Stock Price at Maturity")
    plt.xlabel("Stock Price")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()


    plt.figure(figsize = (10,5))
    plt.hist(payoff, bins = 50, color = 'lightgreen', edgecolor = 'black')
    plt.title(f"{'CALL' if option_type.lower() == 'call' else 'PUT'} Option Payoff Distribution")
    plt.xlabel("Payoff")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()   
    
```

```python
def monte_carlo_option_pricing(S0, K, T, r, sigma, N, option_type, plot = False):
    # Step 1: Simulate terminal prices
    S_T = simulate_gbm_paths(S0, T, r, sigma, N)

    # Step 2: Calculate payoff
    payoff = calculate_payoff(S_T, K, option_type)

    # Step 3: Discount to present value
    price = discount_payoff(payoff, r, T)

    # Step 4: Plot diagnostics (optional but professional)
    if plot:
        plot_histograms(S_T, payoff, option_type)

    return price
```

```python
option_type = 'call'
price = monte_carlo_option_pricing(100, 100, 1, 0.05, 0.2, 10000, option_type, plot = True)
print(f"Estimated {option_type.upper()} Option Price: {price:.4f}")
```

```python
# Plotting Convergence of Monte Carlo Option Price Estimate

simulation_range = np.linspace(1000, 100000, 100, dtype=int)

def plot_convergence(S0, K, T, r, sigma, option_type, simulation_range):
    prices = []

    for N in simulation_range:
        price = monte_carlo_option_pricing(S0, K, T, r, sigma, N, option_type, plot = False)
        prices.append(price)

    plt.figure(figsize=(10,5))
    plt.plot(simulation_range, prices, marker='o', linestyle='-')
    plt.axhline(y=prices[-1], color='red', linestyle='--', label=f"Final Estimate: {prices[-1]:.4f}")
    plt.title("Convergence of Monte Carlo Option Price Estimate")
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Option Prices")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_convergence(S0=100, K=100, T=1, r=0.05, sigma=0.2, option_type='put', simulation_range=simulation_range)
```

```python
# Building Confidence Interval
N = 10000
standard_error = float(np.std(payoff) / np.sqrt(N))

lower_bound = price - 1.96 * standard_error
upper_bound = price + 1.96 * standard_error

print(f"The Estimated European {option_type} Option Price is {price:.4f}\
 with 95% Confidence Interval of [{lower_bound:.4f}, {upper_bound:.4f}]")

plt.figure(figsize = (8,4))
plt.axhline(price, color='blue', linestyle='-', label='Estimated Price')
plt.axhline(lower_bound, color='red', linestyle='--', label='95% confidence interval')
plt.axhline(upper_bound, color='red', linestyle='--')
plt.title("Estimated Option Price with 95% Confidence Interval")
plt.xlabel("Simulation")
plt.ylabel("Option Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Result and Diagnostics
# Comparing Monte Carlo Estimated Price with Black Scholes Analytical Price
# Printing Absolute Difference

from scipy.stats import norm

def black_scholes_price(option_type, S, K, T, r, sigma):
    d1 = ((np.log(S/K)) + (r + ((sigma**2)/2)) * T) / sigma * np.sqrt(T)
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type.lower() == "call":
        price = S * (norm.cdf(d1)) - K * (np.exp(-r * T)) * norm.cdf(d2)

    elif option_type.lower() == "put":
        price = K * (np.exp(-r * T)) * norm.cdf(d2) - S * (norm.cdf(d1))

    else:
        raise ValueError("Invalid option type: must be 'call' or 'put'")

    return float(price)
```

```python
BS_price = black_scholes_price('call', 100, 100, 1, 0.05, 0.2)
MC_price = monte_carlo_option_pricing(100, 100, 1, 0.05, 0.2, 10000, 'call')

print(f"Black Scholes Price: {BS_price:.4f}")
print(f"Monte Carlo Price: {MC_price: .4f}")
print(f"Absolute Difference: {(MC_price - BS_price):.4f}")
```
