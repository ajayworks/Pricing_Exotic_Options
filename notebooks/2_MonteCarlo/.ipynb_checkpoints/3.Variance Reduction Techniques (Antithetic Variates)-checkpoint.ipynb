In this notebook, we explore variance reduction techniques in Monte Carlo simulation for European option pricing. 
Monte Carlo methods are powerful but can be computationally expensive due to high variance. 
We will implement the **Antithetic Variates** technique to reduce variance and improve estimation accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulate Geomatric Brownian Motion for Stock Terminal Price (Stock Price at Maturity)

def simulate_GBM_path(S0, T, r, sigma, N):
    Z = np.random.randn(N)

    S_T = S0 * np.exp((r - 0.5*sigma**2) * T + (sigma * np.sqrt(T) * Z))

    return S_T
```

```python
S_T = simulate_GBM_path(100, 1, 0.05, 0.2, 1000)
```

```python
# Calculate Payoff

def calculate_payoff(S_T, K, option_type):
    if option_type.lower() == 'call':
        payoff = np.maximum(S_T - K, 0)

    elif option_type.lower() == 'put':
        payoff = np.maximum(K - S_T, 0)

    else:
        raise ValueError("Invalid Option Type, Must be 'call' or 'put")

    return payoff
```

```python
payoff = calculate_payoff(S_T, 100, 'call')
```

```python
# Discount Payoff to Present Value (Monte Carlo Estimate of Option Price)

def discount_payoff(payoff, r, T):
    price = np.exp(-r * T) * np.mean(payoff)

    return price

price = discount_payoff(payoff, 0.05, 1)
```

```python
# Monte Carlo Option Pricing

def monte_carlo_option_pricing(S0, K, r, T, sigma, N, option_type):
    S_T = simulate_GBM_path(S0, T, r, sigma, N)
    payoff = calculate_payoff(S_T, K, option_type)
    price = discount_payoff(payoff, r, T)

    return price

MC_price = monte_carlo_option_pricing(100, 100, 0.05, 1, 0.2, 1000, 'call')

print(f"Monte Carlo Estimated Option Price: {MC_price:.4f}")
```

**Antithetic Variates** work by using negatively correlated simulations.
For every standard normal `Z`, we also simulate `-Z`, reducing variance in expected outcomes.

This relies on the symmetry of the normal distribution.

```python
# Implementing Antithetic Variates

def simulate_GBM_antithetic(S0, r, T, sigma, N):
    Z = np.random.randn(N//2)
    Z_full = np.concatenate([Z, -Z])

    S_T = S0 * np.exp((r - 0.5 * sigma**2) * T + (sigma * np.sqrt(T) * Z_full))

    return S_T
```

```python
S_T = simulate_GBM_antithetic(100, 0.05, 1, 0.2, 1000)
```

```python
# Price using Antithetic Simulation

def monte_carlo_antithetic_pricing(S0, K, r, T, sigma, N, option_type):
    S_T = simulate_GBM_antithetic(S0, r, T, sigma, N)
    payoff = calculate_payoff(S_T, K, option_type)
    price = discount_payoff(payoff, r, T)

    return price

price = monte_carlo_antithetic_pricing(100, 100, 0.05, 1, 0.2, 1000, 'call')

print(f"Monte Carlo Antithetic Option Price: {price:.4f}")
```

```python
# Comparing Standard Monte Carlo Price with Anithetic Price

S0, K, r, T, sigma, N = 1000, 1000, 0.05, 1, 0.2, 100000

std_price = monte_carlo_option_pricing(S0, K, r, T, sigma, N, 'call')
anti_price = monte_carlo_antithetic_pricing(S0, K, r, T, sigma, N, 'call')

print(f"Standard MC Price: {std_price:.4f}")
print(f"Antithetic MC Price: {anti_price:.4f}")
```

We observe that the antithetic variates technique maintains the accuracy of the Monte Carlo estimate, 
while potentially reducing the variance in estimation. This is useful in building robust pricing models 
with fewer simulations.

```python
# Antithetic Convergence Plot

simulation_range = np.linspace(1000, 100000, 100, dtype=int)

def antithetic_convergence_plot(S0, K, r, T, sigma, option_type, Simulation_range):
    prices = []

    for N in simulation_range:

        price = monte_carlo_antithetic_pricing(S0, K, r, T, sigma, N, option_type)
        prices.append(price)

    plt.figure(figsize=(8,5))
    plt.plot(simulation_range, prices, marker='o', linestyle='-')
    plt.axhline(y=prices[-1], color='red', linestyle='--', label=f"Final Estimate: {prices[-1]:.4f}")
    plt.title("Antithetic Convergence Plot")
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Option Prices")
    plt.legend()
    plt.grid(True)
    plt.show()     
```

```python
antithetic_convergence_plot(S0=100, K=100, r=0.05, T=1, sigma=0.2, option_type='call', Simulation_range=simulation_range)
```

```python
# Comparing the Varience (Standard Deviation) of both the methods
# Compute same number of simulation paths multiple times.

def comparing_varience(S0, K, r, T, sigma, N, option_type, num_trials = 100):
    mc_prices = []
    anti_prices = []

    for i in range(num_trials):
        mc_price = monte_carlo_option_pricing(S0, K, r, T, sigma, N, option_type)
        anti_price = monte_carlo_antithetic_pricing(S0, K, r, T, sigma, N, option_type)

        mc_prices.append(mc_price)
        anti_prices.append(anti_price)

    sd_mc_prices = np.std(mc_prices)
    sd_anti_prices = np.std(anti_prices)

    print(f"Standard Monte Carlo Std Dev: {sd_mc_prices:.4f}")
    print(f"Antithetic Variates Std Dev: {sd_anti_prices:.4f}")
    print(f"Varience Reduction Ratio: {(sd_mc_prices**2 / sd_anti_prices**2):.4f}")
```

```python
comparing_varience(S0=100, K=100, r=0.05, T=1, sigma=0.2, N=10000, option_type='call')
```
