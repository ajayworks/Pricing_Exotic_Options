```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

 # S0: initial stock price
 # mu: expected return (drift)
 # sigma: volatility (annual)
 # T: time in years
 # N: number of time steps
 # M: Number of simulations (Paths)

def simulate_GBM_paths(S0, mu, sigma, T, N, M):
    dt = T/N   # dt: time step size
    Z = np.random.normal(0, 1, (M, N))
    S = np.zeros((M, N+1))
    S[:, 0] = S0

    for t in range(1, N+1):
        S[:, t] = S[:, t-1] * np.exp((mu-0.5 * sigma**2)*dt + (sigma * np.sqrt(dt) * Z[:, t-1])) #GBM discretized formula

    return S
```

```python
S = simulate_GBM_paths(100, 0.1, 0.2, 1, 252, 10)
```

```python
plt.figure(figsize = (10,6))

for i in range(S.shape[0]):
    plt.plot(S[i])

plt.title("Simulated GBM Paths for a Stock")
plt.xlabel("Time Steps (Days)")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```
