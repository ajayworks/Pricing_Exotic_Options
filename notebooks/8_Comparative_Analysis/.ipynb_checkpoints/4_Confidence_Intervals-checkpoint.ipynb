# Confidence Intervals in Monte Carlo Option Pricing

We estimate a European call price via Monte Carlo and construct
a 95 % confidence interval. We then compare it to the closed‐form
Black–Scholes price.

| Symbol   | Meaning             | Value      |
|:--------:|:-------------------:|:----------:|
| \(S_0\)   | Spot price          | 100        |
| \(K\)     | Strike              | 100        |
| \(r\)     | Risk-free rate      | 5 %        |
| $(\sigma)$| Volatility          | 20 %       |
| \(T\)     | Maturity (years)    | 1.0        |
| Paths     | MC simulation count | 200 000    |
| Steps     | Time‐steps per path | 1 (terminal) |

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt

# allow imports from pricing
sys.path.append(os.path.abspath("../.."))

from pricing.bsm import bsm_price

# Model parameters
S0, K, r, sigma, T = 100, 100, 0.05, 0.20, 1.0
N_paths            = 200_000
```

```python
# Simulate terminal payoffs under Black–Scholes GBM
np.random.seed(42)
Z       = np.random.randn(N_paths)                          # standard normals
ST      = S0 * np.exp((r - 0.5*sigma**2) * T + sigma*np.sqrt(T) * Z)
payoffs = np.exp(-r*T) * np.maximum(ST - K, 0.0)             # discounted call payoff

# Monte Carlo estimate and 95% CI
price_mc = payoffs.mean()
stderr   = payoffs.std(ddof=1) / np.sqrt(N_paths)
ci95     = 1.96 * stderr

# Black–Scholes analytic price
price_bs = bsm_price(S0, K, r, sigma, T, is_call=True)

# Print results

print(f"Monte Carlo estimate  = {price_mc:.4f}  ± {ci95:.4f}  (95 % CI)")
print(f"Black–Scholes price   = {price_bs:.4f}")
```

```python
plt.figure(figsize=(8,5))

# Histogram of the discounted payoffs
plt.hist(payoffs, bins=50, density=True,
         alpha=0.7)

# MC mean
plt.axvline(price_mc, color='red', linestyle='-',
            linewidth=2, label=f"MC mean = {price_mc:.4f}")

# 95% CI bounds
lower, upper = price_mc - ci95, price_mc + ci95
plt.axvline(lower, color='green', linestyle='--',
            label=f"95 % CI = [{lower:.4f}, {upper:.4f}]")
plt.axvline(upper, color='green', linestyle='--')

# Analytic BSM price
plt.axvline(price_bs, color='blue', linestyle='-.',
            linewidth=2, label=f"BSM price = {price_bs:.4f}")

plt.xlabel('Discounted Payoff')
plt.ylabel('Density')
plt.title('Distribution of MC Payoffs with 95 % CI & BSM Price')
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Summary

- Monte Carlo yields **\({price_mc:.4f} \pm {ci95:.4f}\)** at 95 % confidence,  
  which **contains** the analytic BSM price \({price_bs:.4f}\).  
- The histogram visualizes the sampling distribution of discounted payoffs.  
- MC error decays as \(O(N^{-1/2})\); here \(N=200{,}000\) gives ~±0.003 accuracy.  
- For production or Greeks, deterministic PDE or analytic formulas are preferred for higher precision.
