# Time-Complexity & Accuracy Scaling

We compare how runtime and accuracy (error vs Black–Scholes) scale  
with resolution for three vanilla-call pricing methods:

| Symbol        | Meaning                       | Value / Range                        |
|:-------------:|:-----------------------------:|:------------------------------------:|
| \(S_0\)       | Initial spot price            | **100**                              |
| \(K\)         | Strike price                  | 100                                  |
| \(r\)         | Risk-free rate                | 5 %                                  |
| $(\sigma)$    | Volatility (annual)           | 20 %                                 |
| \(T\)         | Time to maturity (years)      | 1.0                                  |
| **BSM**       | Analytic reference price      | computed below                       |
| **Binomial**  | Steps \(N\)                   | 50,150,250,…,950                     |
| **Monte Carlo** | Paths \(N\)                 | $(10^3,10^4,…,10^5)$                 |
| **PDE**       | Grid \(N_S=N_t=N\)            | 20,40,80,160,320                     |

```python
import numpy as np
import matplotlib.pyplot as plt
import time
import os, sys

sys.path.append(os.path.abspath("../../"))

from pricing.pde           import crank_nicolson
from pricing.monte_carlo   import mc_european_price
from pricing.binomial_tree import binomial_crr_price
from pricing.bsm           import bsm_price
```

```python
# Parameters
S0, K, r, sigma, T = 100, 100, 0.05, 0.20, 1.0
is_call            = True

# Resolution ranges
binomial_steps = np.arange(50, 1000, 100)                # 50,150,...,950
mc_paths       = np.logspace(3, 5, 9, dtype=int)         # 1e3 → 1e5
pde_steps      = np.array([20, 40, 80, 160, 320])        # grid sizes

# Analytic Black-Scholes price
price_bs = bsm_price(S0, K, r, sigma, T, is_call=is_call)
print(f"Black–Scholes reference price = {price_bs:.4f}")
```

```python
# -- Binomial tree timing & error --
bin_times  = []
bin_errors = []

for N in binomial_steps:
    start = time.time()
    p = binomial_crr_price(S0, K, r, sigma, T, N, is_call=is_call)
    bin_times.append(time.time() - start)
    bin_errors.append(abs(p - price_bs))

# -- Monte Carlo timing & error --
mc_times  = []
mc_errors = []

for N in mc_paths:
    start = time.time()
    p, ci95 = mc_european_price(
        S0, K, r, sigma, T,
        N_paths=N, N_steps=1,
        is_call=is_call, seed=42
    )
    mc_times.append(time.time() - start)
    mc_errors.append(abs(p - price_bs))

# -- PDE Crank–Nicolson timing & error --
pde_times  = []
pde_errors = []

for N in pde_steps:
    start = time.time()
    p = crank_nicolson(
        S0, K, r, sigma, T,
        Smax=4*K, N_S=N, N_t=N,
        is_call=is_call
    )
    pde_times.append(time.time() - start)
    pde_errors.append(abs(p - price_bs))
```

```python
# Price & error arrays
pde_errors      = [abs(crank_nicolson(S0,K,r,sigma,T,4*K,N,N,True) - price_bs)  for N in pde_steps]
mc_errors       = [abs(mc_european_price(S0,K,r,sigma,T,N,1,True,42)[0] - price_bs) for N in mc_paths]
binomial_errors = [abs(binomial_crr_price(S0,K,r,sigma,T,N,True) - price_bs)       for N in binomial_steps]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Panel 1: Runtime scaling (log-log)
ax1.loglog(binomial_steps, bin_times,  's-', label='Binomial')
ax1.loglog(mc_paths,       mc_times,    'o-', label='Monte Carlo')
ax1.loglog(pde_steps,      pde_times,   'd-', label='PDE (Crank–Nicolson)')
ax1.set_xlabel('Resolution (log scale)')
ax1.set_ylabel('Runtime (s, log scale)')
ax1.set_title('Time Complexity')
ax1.grid(which='both', ls=':', alpha=0.5)
ax1.legend()

# Panel 2: Accuracy vs runtime (log-log)
ax2.loglog(bin_times,  bin_errors,  's-', label='Binomial')
ax2.loglog(mc_times,   mc_errors,   'o-', label='Monte Carlo')
ax2.loglog(pde_times,  pde_errors,  'd-', label='PDE')
ax2.set_xlabel('Runtime (s, log scale)')
ax2.set_ylabel('Absolute error vs BSM (log scale)')
ax2.set_title('Runtime vs Accuracy')
ax2.grid(which='both', ls=':', alpha=0.5)
ax2.legend()

plt.tight_layout()

# save the plot into plots folder
plt.savefig("../../plots/error_runtime_comparison.png", dpi=300)

plt.show()
```

### Observations

- **Binomial-CRR** runtime ∝ $(O(N))$ (linear in steps), error ∝ $(O(N^{-1}))$.  
- **Monte Carlo** runtime ∝ $(O(P))$ (linear in paths), error ∝ $(O(P^{-1/2}))$.  
- **PDE (Crank–Nicolson)** runtime ∝ $(O(N^2))$ (grid × time), error ∝ $(O(N^{-2}))$.

**Conclusion:**  
For pure vanilla calls, **PDE** offers the best accuracy per second at moderate resolutions.  
**Monte Carlo** is flexible but slow to reduce sampling noise.  
**Binomial trees** strike a simple trade-off for teaching and moderate-precision tasks.
