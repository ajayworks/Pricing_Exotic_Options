# Confidence Intervals in Monte Carlo Option Pricing
This notebook demonstrates how to compute and interpret confidence intervals in Monte Carlo pricing of European options.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import os, sys

sys.path.append(os.path.abspath("../../"))

from pricing.monte_carlo import MonteCarloPricer
from pricing.bsm import black_scholes_price
```

```python
# Option and model parameters
S0 = 100
K = 100
T = 1.0
r = 0.05
sigma = 0.2
option_type = 'call'

n_paths = 10000  
confidence = 0.95
```

```python
# Pricing function that also returns confidence interval
mc_pricer = MonteCarloPricer(S0=S0, K=K, T=T, r=r, sigma=sigma, n_paths=n_paths, option_type=option_type)

price, (ci_lower, ci_upper) = mc_pricer.price_with_confidence_interval()
ref_price = black_scholes_price(S0, K, T, r, sigma, option_type)

print(f"Monte Carlo Price: {price:.4f}")
print(f"95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Black-Scholes Reference: {ref_price:.4f}")
```

### ðŸŽ¯ Interpretation

- The Monte Carlo estimate is subject to sampling noise.
- The 95% confidence interval provides a statistical range where the true price is expected to lie.
- The BSM price falling within the interval validates the implementation.

```python
# Show payoff distribution and confidence interval
payoffs = mc_pricer.simulate_payoffs()
discounted = np.exp(-r * T) * payoffs

plt.hist(discounted, bins=50, alpha=0.7)
plt.axvline(price, color='blue', label='MC Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='Lower CI')
plt.axvline(ci_upper, color='green', linestyle='--', label='Upper CI')
plt.axvline(ref_price, color='red', linestyle=':', label='BSM Price')

plt.title("Distribution of Discounted Payoffs")
plt.xlabel("Payoff")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
