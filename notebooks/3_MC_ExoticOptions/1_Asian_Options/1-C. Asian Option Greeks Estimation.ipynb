# Asian Option Greeks Estimation

This notebook focuses on estimating the Greeks (Delta, Gamma, Vega) for an arithmetic average Asian call option using Monte Carlo simulation.

We use **finite difference approximations** with **common random numbers (CRN)** to stabilize variance. This analysis is essential for hedging and risk management of path-dependent exotic derivatives.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a fixed seed for reproducibility
np.random.seed(42)

# Parameters
S0 = 100      # Initial stock price
K = 100       # Strike price
r = 0.05      # Risk-free rate
T = 1.0       # Time to maturity
sigma = 0.2   # Volatility
M = 100       # Time steps
N = 100000    # Number of simulations
h = 0.01      # Bump size for finite difference
```

```python
# Monte Carlo Price of Asian Call

def simulate_paths(S0, r, sigma, T, M, N, Z=None):
    dt = T / M
    if Z is None:
        Z = np.random.randn(N, M)
    S = np.zeros((N, M+1))
    S[:, 0] = S0
    for t in range(1, M+1):
        S[:, t] = S[:, t-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z[:, t-1])
    return S

def asian_call_price(S):
    average_price = np.mean(S[:, 1:], axis=1)  # exclude initial price
    payoff = np.exp(-r * T) * np.maximum(average_price - K, 0)
    return np.mean(payoff)
```

```python
# Greeks Estimation (Finite Difference + CRN)

def estimate_asian_greeks(S0, K, r, T, sigma, M, N, h):
    Z = np.random.randn(N, M)  # Common random numbers for all bumps

    # Base
    S_base = simulate_paths(S0, r, sigma, T, M, N, Z)
    price_base = asian_call_price(S_base)

    # Delta
    S_up = simulate_paths(S0 + h, r, sigma, T, M, N, Z)
    S_down = simulate_paths(S0 - h, r, sigma, T, M, N, Z)
    delta = (asian_call_price(S_up) - asian_call_price(S_down)) / (2 * h)

    # Gamma
    gamma = (asian_call_price(S_up) - 2 * price_base + asian_call_price(S_down)) / (h ** 2)

    # Vega
    S_vega_up = simulate_paths(S0, r, sigma + h, T, M, N, Z)
    S_vega_down = simulate_paths(S0, r, sigma - h, T, M, N, Z)
    vega = (asian_call_price(S_vega_up) - asian_call_price(S_vega_down)) / (2 * h)

    return price_base, delta, gamma, vega
```

```python
price, delta, gamma, vega = estimate_asian_greeks(S0, K, r, T, sigma, M, N, h)

print(f"Base Price : {price:.4f}")
print(f"Delta      : {delta:.4f}")
print(f"Gamma      : {gamma:.4f}")
print(f"Vega       : {vega:.4f}")
```

```python
S_range = np.linspace(80, 120, 15)
deltas, gammas = [], []

for S in S_range:
    _, delta, gamma, _ = estimate_asian_greeks(S, K, r, T, sigma, M, N, h)
    deltas.append(delta)
    gammas.append(gamma)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(S_range, deltas, label='Delta', marker='o')
plt.title('Delta vs S₀')
plt.xlabel('S₀'); plt.ylabel('Delta'); plt.grid(); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(S_range, gammas, label='Gamma', marker='o', color='green')
plt.title('Gamma vs S₀')
plt.xlabel('S₀'); plt.ylabel('Gamma'); plt.grid(); plt.legend()
plt.tight_layout()
plt.show()
```

```python
sigma_range = np.linspace(0.1, 0.5, 15)
vegas = []

for sig in sigma_range:
    _, _, _, vega = estimate_asian_greeks(S0, K, r, T, sig, M, N, h)
    vegas.append(vega)

plt.figure(figsize=(6, 4))
plt.plot(sigma_range, vegas, label='Vega', marker='o', color='purple')
plt.title('Vega vs σ')
plt.xlabel('Volatility (σ)'); plt.ylabel('Vega')
plt.grid(); plt.legend()
plt.tight_layout()
plt.show()
```

### Visual Insights

- **Delta** increases with \( S_0 \) as expected, showing the option becomes more sensitive to price.
- **Gamma** peaks near the strike price — typical for vanilla and path-dependent options.
- **Vega** increases with volatility, suggesting more value from uncertain paths.

These visuals confirm that our finite-difference Greeks behave consistently with financial intuition, despite being computed via Monte Carlo simulation.
